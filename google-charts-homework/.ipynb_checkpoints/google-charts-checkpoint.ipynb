````python
import pandas as pd
import numpy as np
````

````python
# load GDP Data 
# change to Billion by / 10
gdp = pd.read_excel("data/AnnualbyProvince-gdp.xls", skiprows=3, nrows=31, index_col=0)
gdp = (gdp / 10).round(2)
gdp.head()
````

# Google Bar Chart

````python
df = gdp["2022"].sort_values(ascending=False).reset_index()
df['annotation'] = df['2022'].apply(lambda x: f"{x:,.0f}")
df.values
````

# Line Chart

````python
# load GDP Data 
# change to Billion by / 10
gdp = pd.read_excel("data/Annual_20years.xls", skiprows=2, nrows=1, index_col=0)
gdp = (gdp / 10).round(2)
gdp.head()
````

````python
gdp.T.reset_index().sort_values('index').values
````

# Treemap Chart

````python
# load GDP Data 
# change to Billion by / 10
gdp = pd.read_excel("data/AnnualbyProvince-gdp.xls", skiprows=3, nrows=31, index_col=0)
gdp = (gdp / 10).round(2)[["2022"]]
gdp['Parent'] = 'China'
gdp['size'] = gdp['2022']
gdp.reset_index()[['Region', 'Parent', '2022', 'size']].values
````

# Bubble Chart

````python
life = pd.read_excel("data/AnnualbyProvince-life.xls", skiprows=3, nrows=31, index_col=0)
life.columns  = ['life']
population = pd.read_excel("data/AnnualbyProvince-population.xls", skiprows=3, nrows=31, index_col=0)
population.columns  = ['population']
rate = pd.read_excel("data/AnnualbyProvince-birthrate.xls", skiprows=3, nrows=31, index_col=0)
rate.columns  = ['birth_rate']
df = pd.concat([rate, population, life], axis=1)
df['parent'] = 'China'
df.reset_index()[['Region', 'life', 'birth_rate', 'parent', 'population']].values
````
